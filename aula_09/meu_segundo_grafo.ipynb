```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

nós = ['Audrey', 'Bill', 'Cathy', 'Dave', 'Emily', 'Frank', 'Gayle', 'Harry', 'Ilana', 'Jeff']
G.add_nodes_from(nós)
```

```python
conexoes = [
    ('Audrey', 'Bill'), ('Audrey', 'Gayle'), ('Audrey', 'Ilana'),
    ('Bill', 'Emily'), ('Bill', 'Gayle'), ('Bill', 'Jeff'),
    ('Cathy', 'Emily'), ('Cathy', 'Gayle'),
    ('Dave', 'Ilana'), ('Dave', 'Emily'), ('Dave', 'Frank'),
    ('Emily', 'Bill'), ('Emily', 'Cathy'), ('Emily', 'Jeff'), ('Emily', 'Dave'), ('Emily', 'Frank'),
    ('Frank', 'Dave'), ('Frank', 'Emily'),
    ('Gayle', 'Audrey'), ('Gayle', 'Bill'), ('Gayle', 'Cathy'),
    ('Harry', 'Jeff'), ('Harry', 'Ilana'),
    ('Ilana', 'Harry'),  ('Ilana', 'Dave'),
    ('Jeff', 'Harry'), ('Jeff', 'Bill'),
]

G.add_edges_from(conexoes)
```

```python
pos = nx.spring_layout(G, seed=105)
nx.draw(G, pos, with_labels=True, node_color='lightgreen',
        node_size=1200, font_weight='bold', font_size=9)

plt.title('Representação - Atividade 1')
plt.show()

print(f"Total de Nós: {G.number_of_nodes()}")
print(f"Total de Arestas: {G.number_of_edges()}")
```
