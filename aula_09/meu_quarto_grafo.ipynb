```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_nodes_from(['Prince', 'David Bowie', 'Madonna', 'Lady Gaga', 'Michael Jackson', 'Beyoncé', 'Kanye West', 'Rihanna', 'Jay-Z', ])
```

```python
G.add_edges_from([('Prince', 'Michael Jackson'),
                  ('David Bowie', 'Prince',),
                  ('David Bowie', 'Madonna',),
                  ('Michael Jackson', 'Beyoncé'),
                  ('Beyoncé', 'Rihanna'),
                  ('Kanye West', 'Rihanna'),
                  ('Jay-Z', 'Rihanna',),
                  ('Jay-Z', 'Kanye West'),
                  ('Madonna', 'Lady Gaga'),])

print(f"Nós do Grafo : {G.nodes()}")
print(f"Arestas do Grafo: {G.edges()}")
```

```python
plt.figure(figsize=(10, 7))

pos = nx.spring_layout(G, k=1.2, seed =52)

nx.draw(G, pos, with_labels=True, node_color='red',
        node_size=2000, font_color='white', font_size=7, arrowsize=15)

plt.title('Representação em Grafo Direcionado')
plt.show()
```
