```python
import networkx as nx  # BIBLIOTECA PARA GRAFOS
import matplotlib.pyplot as plt #IMPRIMIR GRAFOS

# Grafo Orientado
G = nx.DiGraph()

# Adicionar os Nós / Vértices
G.add_nodes_from(['Jailson Mendes', 'Fã-1', 'Fã-2', 'Fã-3', 'Não-Fã'])
```

```python
# Adicionar Arestas
G.add_edges_from([('Fã-1', 'Jailson Mendes'),
                  ('Fã-2', 'Jailson Mendes'),
                  ('Fã-3', 'Jailson Mendes'),])

print(f"Nós do Grafo : {G.nodes()}")
print(f"Arestas do Grafo: {G.edges()}")
```

```python
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels=True, node_color='red',
        node_size=2000, font_color='white', font_size=7, arrowsize=15)

plt.title('Representação em Grafo Direcionado')
plt.show()
```
